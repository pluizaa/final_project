```python
import pickle
import dash
import dash_core_components as dcc
import dash_html_components as html
import dash_daq as daq
from dash.dependencies import Input, Output
import dash_bootstrap_components as dbc
import numpy as np
import pandas as pd
```

```python
with open('tree_model.pickle', 'rb') as f:
    ccc = pickle.load(f)
```

```python
app = dash.Dash()

# The page structure will be:
#    Features Importance Chart
#    <H4> Feature #1 name
#    Slider to update Feature #1 value
#    <H4> Feature #2 name
#    Slider to update Feature #2 value
#    <H4> Feature #3 name
#    Slider to update Feature #3 value
#    <H2> Updated Prediction
#    Callback fuction with Sliders values as inputs and Prediction as Output

with open('tree_model.pickle', 'rb') as f:
    ccc = pickle.load(f)
    
# We apply basic HTML formatting to the layout
app.layout = html.Div(style={'textAlign': 'center', 'font-family': 'Verdana','background-image': 'url(https://aaaliving.acg.aaa.com/wp-content/uploads/2019/11/debit-vs-credit-card-holiday-shopping.gif)',
                            'background-position': 'right top','background-repeat': 'no-repeat', 'background-size': '30%',
                             'background-color': '#FAE8A6'},
                    
                      children=[
                        

                        # Title display
                        html.H1(children="Default Prediction Tool"),
                           
                        html.H4(children='Revolving Utilization Of Unsecured Lines'),
                        dcc.Input(id='RevolvingUtilizationOfUnsecuredLines', value='input value', type='number'),
                        
                        html.H4(children='Age'),
                        dcc.Input(id='age', value='input value', type='number'),   
                        
                        html.H4(children='Number Of Time 30-59 Days Past Due Not Worse'),
                        dcc.Input(id='NumberOfTime30_59DaysPastDueNotWorse', value='input value', type='number'),
                             
                        html.H4(children='Debt Ratio'),
                        dcc.Input(id='DebtRatio', value='input value', type='number'),
                             
                        html.H4(children='Monthly Income'),
                        dcc.Input(id='MonthlyIncome', value='input value', type='number'),
                             
                        html.H4(children='Number Real Estate Loans Or Lines'),
                        dcc.Input(id='NumberRealEstateLoansOrLines', value='input value', type='number'),
                         
                        html.H4(children='Number Of Open Credit Lines And Loans'),
                        dcc.Input(id='NumberOfOpenCreditLinesAndLoans', value='input value', type='number'),
                             
                        html.H4(children='Number Of Times 90 Days Late'),
                        dcc.Input(id='NumberOfTimes90DaysLate', value='input value', type='number'),
                             
                        html.H4(children='Number Of Time 60-89 Days Past Due Not Worse'),
                        dcc.Input(id='NumberOfTime60_89DaysPastDueNotWorse', value='input value', type='number'),
                       
                        html.H4(children='Number Of Dependents'),
                        dcc.Input(id='NumberOfDependents', value='input value', type='number'),
                        
                
                        html.H2(id='prediction_result'),
                         

                    ])
```

```python
@app.callback(Output(component_id="prediction_result",component_property="children"),
# The values correspnding to the three sliders are obtained by calling their id and value property
              [Input("RevolvingUtilizationOfUnsecuredLines","value"), 
               Input("age","value"),
               Input("NumberOfTime30_59DaysPastDueNotWorse","value"),
               Input("DebtRatio","value"),
               Input("MonthlyIncome","value"),
               Input("NumberRealEstateLoansOrLines","value"),
               Input("NumberOfOpenCreditLinesAndLoans","value"),
               Input("NumberOfTimes90DaysLate","value"),
               Input("NumberOfTime60_89DaysPastDueNotWorse","value"),
               Input("NumberOfDependents","value")
              ])

def update_output(RevolvingUtilizationOfUnsecuredLines,age,NumberOfTime30_59DaysPastDueNotWorse,
                  DebtRatio,MonthlyIncome, NumberRealEstateLoansOrLines, NumberOfOpenCreditLinesAndLoans,
                  NumberOfTimes90DaysLate,NumberOfTime60_89DaysPastDueNotWorse,NumberOfDependents):    
    
    x = np.array([[RevolvingUtilizationOfUnsecuredLines,age,NumberOfTime30_59DaysPastDueNotWorse,DebtRatio,MonthlyIncome  ,
                   NumberRealEstateLoansOrLines, NumberOfOpenCreditLinesAndLoans,NumberOfTimes90DaysLate,
                   NumberOfTime60_89DaysPastDueNotWorse,NumberOfDependents ]])
    
    prediction = ccc.predict(x)[0]
   # And retuned to the Output of the callback function
    if prediction == 0:
        output = 'NO default'
    else:
        output = 'Default'
    return f'The predicted outcome is: {output}.'

### Run the App ###############################################

if __name__ == "__main__":
    app.run_server()
    
```
